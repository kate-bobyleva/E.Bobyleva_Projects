# Прогноз оттока клиентов из сети фитнес-центров

## Описание проекта

Есть сеть фитнес-центров, где ставят задачу по борьбе с оттоком клиентов. Для этого предлагается проанализировать характеристики клиентов, их поведение в отношении посещения и пользования услугами фитнес-клубов. В нашем распоряжении информация по нескольким клиентам, которая содержит данные на месяц до оттока и факт оттока на определённый месяц.

**Цель проекта** - провести анализ и подготовить рекомендации по удержанию клиентов.

**Задачи:**
* научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
* сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
* проанализировать основные признаки, наиболее сильно влияющие на отток;
* сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:

    1) выделить целевые группы клиентов;
    
    2) предложить меры по снижению оттока;
    
    3) определить другие особенности взаимодействия с клиентами.

**Обзор выполненных работ:**

I. Предобработка и исследовательский анализ имеющихся данных.

II. Построение модели прогнозирования оттока клиентов

	1. Построение моделей на основе двух алгоритмов.
	2. Оценка метрик моделей и выбор лучшей модели для прогнозирования.

III. Выделение и описание наиболее ярких групп клиентов (кластеризация).

IV. Выводы и базовые рекомендации.



**Описание данных**:

Таблица _gym_churn_ с данными:

Данные клиента за предыдущий до проверки факта оттока месяц:

    'gender' — пол;
    'Near_Location' — проживание или работа в районе, где находится фитнес-центр;
    'Partner' — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);
    Promo_friends — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);
    'Phone' — наличие контактного телефона;
    'Age' — возраст;
    'Lifetime' — время с момента первого обращения в фитнес-центр (в месяцах).
    
Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:

    'Contract_period' — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
    'Month_to_end_contract' — срок до окончания текущего действующего абонемента (в месяцах);
    'Group_visits' — факт посещения групповых занятий;
    'Avg_class_frequency_total' — средняя частота посещений в неделю за все время с начала действия абонемента;
    'Avg_class_frequency_current_month' — средняя частота посещений в неделю за предыдущий месяц;
    'Avg_additional_charges_total' — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.
    'Churn' — факт оттока в текущем месяце.

### Распределение признаков в разрезе групп по факту оттока

Здесь рассмотрим какова плотность распределения признаков в группах оставшихся клиентов и ушедших в отток.

```python
cols = df.columns

sns.set(font_scale=1.2)
sns.set_style('whitegrid')
fig, axes = plt.subplots(5, 3, figsize=(18, 20))
axes = axes.ravel()

for col, ax in zip(cols, axes):
    sns.distplot(df[df['churn'] == 0][col], ax=ax, label='Остались')
    sns.distplot(df[df['churn'] == 1][col], ax=ax, label='Ушли')
    ax.set_title(f'Распределение признака {col}', fontsize=14)
    ax.legend()

fig.tight_layout()
plt.show()
```

Распределения подтверждают вышеизложенные выводы.
Также можно сказать что:
* Несмотря на то, что разница в среднем возрасте небольшая, всё-таки ушедшие в отток клиенты, как правило, в среднем моложе оставшихся. Средний возраст ушедших - 27 лет, отсавшихся - 29 лет.
* Те клиенты, которые оплачивают больше дополнительных услуг - менее склонны уйти в отток.
* Среди тех, кто посещает или посетил групповые занятия, количество оставшихся более чем в два раза больше чем те, кто ушёл в отток.
* Среди клиентов, работающих в компаниях-партнёрах, доля ушедших более чем в 2 раза меньше, чем оставшихся.

В данном случае явно видна мультиколлинеарность между признаками "срок действия текущего абонемента" и "срок до окончания текущего действующего абонемента (в месяцах)", а также между средней частотой посещений в неделю за все время с начала действия абонемента и - за прошедший месяц.

Взаимная корреляция признаков нежелательна, т.к. может привести к переобучению модели, поэтому перед построением модели мы удалим по одному в паре таких признаков, чтобы избавиться от мультиколлинеарности.

Что касается признаков наиболее влияющих на отток, то здесь среднюю степень влияния показывают следующие признаки:
* чем меньше времени прошло с момента первого обращения клиента, тем больше вероятность, что он попадёт в отток.
* чем реже клиент приходит на фитнес, тем скорее он попадёт в отток.
* чем старше клиент, тем менее он склонен к оттоку.
* чем меньше срок абонемента клиента - тем больше вероятность, что он пойдёт в отток.

## Построение модели прогнозирования оттока клиентов

В этом разделе построим модели бинарной классификации клиентов, где целевой признак — факт оттока клиента в следующем месяце.

### Удаление мультиколлинеарных признаков

Перед построением модели, удаляем по одному из мультиколлинеарных признаков.

```python
df_new = df.drop(['month_to_end_contract', 'avg_class_frequency_current_month'], axis=1)
```

```python
# проверяем результат
df_new.info()
```

Признаки, дающие мультиколлинеарность удалены, оставляем в распоряжении 11 признаков.

### Деление на обучающую и валидационную выборку

Так как в данных нет временной структуры, делим выборку на обучающую и валидационную случайно. 

```python
# делим данные на признаки (матрица X) и целевую переменную (y)
X = df_new.drop('churn', axis=1)
y = df_new['churn']
```

Чтобы в обучающую и валидационную выборки попали равные доли клиентов ушедших в отток (то есть чтобы алгоритм имел равные возможности для изучения признаков каждого из представленных классов), применим также аргумент stratify.

```python
# разделяем модель на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0,  stratify=y)
```

```python
# обучаем StandartScaler на обучающей выборке
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
# Преобразуем обучающий и валидационные наборы данных
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
```

### Обучение модели двумя способами

Обучим модель на train-выборке двумя способами:

* логистической регрессией,
* случайным лесом.

#### Логистическая регрессия

```python
# задаём алгоритм для нашей модели
lr_model = LogisticRegression(solver='liblinear', random_state=0)

# обучаем модель
lr_model.fit(X_train_st, y_train)
```

```python
# пользуемся обученной моделью, чтобы сделать прогнозы
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:, 1]
```

#### Случайный лес

```python
# зададим алгоритм для новой модели на основе алгоритма случайного леса
rf_model =  RandomForestClassifier(n_estimators = 100, random_state = 0)

# обучаем модель случайного леса
rf_model.fit(X_train_st, y_train)
```

```python
# пользуемся обученной моделью, чтобы сделать прогнозы
rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)[:, 1]
```

### Оценка метрик моделей и выбор лучшей модели

В этом разделе посмотрим насколько эффективно построенные модели предсказывают факт оттока, и выберем лучшую из этих моделей. Оценивать работу моделей будем по таким метрикам:
* доля правильных ответов (из всех прогнозов) - Accuracy,
* точность (доля правильных ответов только среди целевого класса) - Precision,
* полнота (метрика покажет, сколько реальных клиентов, ушедших в отток модель смогла обнаружить) - Recall.

```python
# определим функцию, которая будет выводить наши метрики
def print_all_metrics(y_true, y_pred, title='Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
```

```python
# выведем все метрики для обеих моделей
print_all_metrics(
    y_test,
    lr_predictions,
    title='Метрики для модели логистической регрессии:'
)

print_all_metrics(
    y_test, 
    rf_predictions,
    title='Метрики для модели случайного леса:'
)
```

Среди построенных моделей по всем метрикам лидирует модель логистической регрессии, её метрики больше всего ближе к единице, что характеризует её лучше, чем модель случайного леса.

Для дальнейшей работы по прогнозированию оттока следует взять эту модель (логистическая регрессия).

Если же значения метрик не были бы так однозначны для определения лучшей модели, то можно было бы обратить внимание на более важные для конкретной ситуации метрики. Например, если бы компания не располагает большими ресурсами и готова тратить их на удержание только тех клиентов, которые с максимальной вероятностью уйдут в отток - здесь стоит обратить внимание на ту модель, которая показывает более высокую метрику Precision (точность).  Если же сеть фитнес-центров готова "потратиться" на удержание как можно большего числа клиентов, вне зависимости от того, насколько велика вероятность попадания клиента в отток, если ей важно не пропустить ни одного случая попадания клиента в отток - то тогда обращаем приоритетное внимание на метрику Recall (полноту) и выбираем модель с бОльшим значением этой метрики.
